```python
#criando dataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando as 5 primeiras linhas
df.head()
```

```python
#quantidade de linhas e colunas
df.shape
```

```python
#verificando data types
df.dtypes
```

```python
#qual a receita total?
df["Valor Venda"].sum()
```

```python
#qual o custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])#criando a coluna de custo
```

```python
#verificando coluna de custo
df.head(1)
```

```python
#qual o custo total?
round(df["custo"].sum(), 2)
```

```python
#criando coluna Lucro(Lucro = Receita - Custo)
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
#verificando coluna de Lucro
df.head(1)
```

```python
#total Lucro
round(df["lucro"].sum(), 2)
```

```python
#criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#média do tempo de envio para cada marca?
#extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtypes
```

```python
#média tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#verificando se tem dados faltantes
df.isnull().sum()
```

```python
#Lucro por ano e Marca?
#agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#formatação números flutuantes
pd.options.display.float_format = '{:20,.2f}'.format 
```

```python
#resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#gráfico total de produtos vendidos
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#lucro x mes
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.xlabel("Lucro");
```

```python
#lucro x marca
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.xlabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
#lucro x classe
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.xlabel("Lucro")
plt.xticks(rotation='horizontal');
```
